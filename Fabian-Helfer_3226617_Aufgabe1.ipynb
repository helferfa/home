## Aufgabe 1

Als erstes wird die Datei eingelesen ingelesen


```python
def loadCSV():
    fobj = open("P.csv")
    for line in fobj:
        numbers = str(line).split(';')
        fobj.close()
        return numbers
```

Dann wird unsere MappingFunktion definiert und die Daten auf Fehler überprüft

```python
def f(text):
    result = {}
    try:
        val = int(text)
        result[val] = 1
    except ValueError:
        print("That's not an int!")
    return result
```

Damit bereiten wir die Daten für die reduce Funktion vor

```python
list1 = loadCSV()
res1 = list(map(f, list1))
```

```python
print(res1)
```

Dann definieren wir die reducer Funktion 

```python
def reducerForReduce(lhs, rhs):
    for k, v in rhs.items(): # in case the dict has more than one entry
        if k in lhs:
            lhs[k] += v
        else:
            lhs[k] = v    
    return lhs
```

```python
import functools
res = functools.reduce(reducerForReduce, res1, {})
```

```python
print(res)
```

Soriterte Ausgabe:

```python
def postProcess(words):
    for k in sorted(words.keys()):
        print(k, words[k])
```

```python
postProcess(res)
```

Bringe die Daten in eine sortierte Form

```python
sorted_res = {}
for k in sorted(res.keys()):
    sorted_res[k] = res[k]
```

Darauf folgt die Diagramm Auswertung


```python
import matplotlib.pyplot as plt


plt.figure(figsize=(100,50))
plt.bar(range(len(sorted_res)), sorted_res.values(), align='center')
plt.xticks(range(len(sorted_res)), sorted_res.keys(), rotation=90)
plt.show()
```
