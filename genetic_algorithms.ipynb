```python
class eight_queens_problem:                
    def __init__(self, state=None):
        self.state = state if state is not None else ''.join(str(randrange(8)) for _ in range(8))
        self.fitness = 28
        self.calc_fitness()
        
    def calc_fitness(self):
        # Check for queens collision in row
        for i in range(len(self.state)):
            for j in range(i+1, len(self.state)):
                if self.state[i] == self.state[j]:
                    self.fitness = self.fitness -1
            index = 0
            for k in range(i+1, len(self.state)):
                index = index +1
                if int(self.state[k]) == int(self.state[i]) + index:
                    self.fitness = self.fitness -1
                if int(self.state[k]) == int(self.state[i]) - index:
                    self.fitness = self.fitness -1
    
    def __getitem__(self, i) -> str:
        return self.state[i]
    
    def __str__(self) -> str:
        l = ['┌───┬───┬───┬───┬───┬───┬───┬───┐']
        for i in range(8):
            l.append('│ ' + ' │ '.join(('x') 
                                       if self[j] == str(i) else ' ' for j in range(len(self.state))) + ' │')
            if i < 7:
                l.append('├───┼───┼───┼───┼───┼───┼───┼───┤')
        l.append('└───┴───┴───┴───┴───┴───┴───┴───┘')
        return '\n'.join(l)
    
    def __lt__(self, other):
        return self.fitness > other.fitness
    
    def mutation(self):
        index = randrange(len(self.state))
        new_state = self.state
        new_state = self.state[:index] + str(randrange(8)) + self.state[index + 1:]
        return eight_queens_problem(new_state)
        
        
                
board = eight_queens_problem('02467531')
print(board)
```

```python
from random import randrange, uniform, choice, seed

class Genetic_algorithm:
    def __init__(self, size, mutation_prob):
        self.mutation_prob = mutation_prob
        self.population = sorted( eight_queens_problem() for i in range(size))
        
    def select(self):
        i = randrange(len(self.population))
        return self.population[i]
    
    def mutate(self, parent):
        if uniform(0, 1) < self.mutation_prob:
            return parent.mutation()
        else:
            return parent
    
    def reproduce(self, parent0, parent1):
        n = len(parent0.state)
        c = randrange(n)
        new_state = parent0.state[:c] + parent1.state[c:]
        new_state2 = parent1.state[:c] + parent0.state[c:]
        return [eight_queens_problem(new_state), eight_queens_problem(new_state2)]
    
    def solve(self):
        iteration = 0
        while iteration<100:
            new_population = []
            for i in range(len(self.population)):
                parent0 = self.select()
                parent1 = parent0
                while parent1 == parent0:
                    parent1 = self.select()
                children = self.reproduce(parent0, parent1)
                child1 = self.mutate(children[0])
                child2 = self.mutate(children[1])
                new_population.append(child1)
                new_population.append(child2)
            self.population = sorted(new_population)
            self.population = self.population[:8]
            iteration = iteration +1
            print('Iteration: ' + str(iteration))
            print(self.population[0])
            print('Fitness: ' + str(self.population[0].fitness))
            print('-------------------------------')
            if self.population[0].fitness == 28:
                break
        print(iteration)
        print(self.population[0])
        print(self.population[0].fitness)
        
solution = Genetic_algorithm(8, 0.3)
solution.solve()
```
