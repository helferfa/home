```python
class eight_queens_problem:                
    def __init__(self, state=None):
        self.state = state if state is not None else ''.join(str(randrange(8)) for _ in range(8))
        self.fitness = 28
        self.calc_fitness()
        
    def calc_fitness(self):
        # Check for queens collision in row
        for i in range(len(self.state)):
            for j in range(i+1, len(self.state)):
                if self.state[i] == self.state[j]:
                    self.fitness = self.fitness -1
            index = 0
            for k in range(i+1, len(self.state)):
                index = index +1
                if int(self.state[k]) == int(self.state[i]) + index:
                    self.fitness = self.fitness -1
                if int(self.state[k]) == int(self.state[i]) - index:
                    self.fitness = self.fitness -1
    
    def __getitem__(self, i) -> str:
        return self.state[i]
    
    def __str__(self) -> str:
        l = ['┌───┬───┬───┬───┬───┬───┬───┬───┐']
        for i in range(8):
            l.append('│ ' + ' │ '.join(('x') 
                                       if self[j] == str(i) else ' ' for j in range(len(self.state))) + ' │')
            if i < 7:
                l.append('├───┼───┼───┼───┼───┼───┼───┼───┤')
        l.append('└───┴───┴───┴───┴───┴───┴───┴───┘')
        return '\n'.join(l)
    
    def __lt__(self, other):
        return self.fitness > other.fitness
    
    def mutation(self):
        index = randrange(len(self.state))
        new_state = self.state
        new_state = self.state[:index] + str(randrange(8)) + self.state[index + 1:]
        return eight_queens_problem(new_state)
        
        
                
board = eight_queens_problem('02467531')
print(board)
```

```python
class CSP:
    def __init__(self, variables, constraints):
        self.variables = {var: None for var in variables.keys()}
        self.domain = {var: domain for var, domain in variables.items()}
        self.constraints = constraints
        self.neighbors = {var: [] for var in self}
        for nodes in constraints.keys():
            for i, node in enumerate(nodes):
                for j in nodes[:i]+nodes[i+1:]:
                    if j not in self.neighbors[node]:
                        self.neighbors[node].append(j)
        #print(self.domain)
        
    def __iter__(self):
        for var in self.variables.keys():
            yield var   
        
    def __in__(self, key):
        if isinstance(key, tuple):
            return key in self.constraints.keys()
        else:
            return key in self.variables
        
    def __getitem__(self, key):
        if isinstance(key, tuple):
            return self.constraints[key]
        else:
            return self.variables[key]
        
    def __setitem__(self, key, val):
        self.variables[key] = val
        
    # returns if all queens have been placed
    def complete(self):
        for key in self.variables:
            if self.variables[key] is None:
                return False
            if not self.consistent(key):
                return False
        return True
    
    def consistent(self, var):
        
        for key in self.variables:
            if not self[key] is None:
                if self[key] == self[var] and not key == var:
                    return False
                if not key==var:
                    #print('Wert' + str(key) + '=' + str(self.variables[key]) + 'Eingefügter Wert: ' + str(var) + '=' + str(self.variables[var]))
                    if int(self.variables[key])+(int(var)-int(key)) == int(self.variables[var]):
                        return False
                    if int(self.variables[key])-(int(var)-int(key)) == int(self.variables[var]):
                        return False
        return True
        
variables  = {idx: list(range(8)) for idx in range(8)}
constraints = {
    (i, j): lambda Xi, Xj, Di, Dj: abs(Xi - Xj) != abs(Di - Dj) and Di != Dj
    for i in range(8) for j in range(8) if i != j
}


def select(csp):
    for column in csp.variables.keys():
        if csp.variables[column] is None:
            return column
        
        
def Backtracking_search(csp, assignment, order_domain=lambda x: x):
    if csp.complete():
        return csp
    # select first column to be assigned
    #print(csp.variables)
    var = select(csp)
    #print('Selected: ' + str(var))
    # für alle annehmbaren werte
    for val in csp.domain[var]:
        #print(csp.domain)
        #print('val=: ' + str(val))
        # wenn keine threats
        csp[var] = val
        if csp.consistent(var):
            # speichere wert
            #print('Eingefügt -> next')
            result = Backtracking_search(csp, assignment)
            if result:
                #print(result.variables)
                liste.append(list(csp.variables.values()))
                s = ''
                for i in range(8):
                    s = s + str(list(csp.variables.values())[i])
                print(s)
                print(eight_queens_problem(s))
        csp.variables[var] = None
    #print('Gelöscht backtracking: ' + str(var))
    return False
        
CSP = CSP(variables, constraints)
liste = []
CSP_result = Backtracking_search(CSP, 0)
print(str(len(liste)) + ' Lösungen gefunden!')
```
